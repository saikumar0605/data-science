```python
dataset.isnull().sum() #it shows wether true if present else false not present.
```

```python
dataset.isnull().sum()  #we will fet missing values acording to the column.
```

```python
#We will perform list comphrension now.
features_with_nan = [features for features in dataset.columns if dataset[features].isnull().sum()>=1]
#here we are getting only the fetures columns which has more than 1 null values. and forming the columns list.
```

```python
features_with_nan
```

```python
#one way to get % of nan velues present in column is.

#dataset['BsmtFinType2'].isnull().mean()

#other intresting way would be.
for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean()*100,4), '% missing values')
#here we get null values of all features/ columns avaliable.
#4 we gave because we want it to roundoff to four decimal values.
```

```python
sns.boxplot(dataset['SalePrice'])
#here we can see many outliers in the data graph. so we used median below to avoid them.
```

```python
#lets  plot and see which are dependent and independent feature..
#if nan value then its 1
#if no nan than its 0

for feature in features_with_nan:
    data = dataset.copy()
    #ABOVE WE COPYED THE DATASET TO DATA.
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    #let's calculate the mean saleprice where the information is missing or present. and we use median to avoide outliers in data.
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

here with the relation between the missing values and the dependent variable is clearly visible. so we need to replace these these nan values with some meaningful values. which we will do in the Feature Engineering section

# Numerical variables.


```python
## list of numerical variables

dataset['SalePrice'].dtypes != 'object' #to check wether it's object or not.
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
## Temporal Variables(Eg: datetime variables.)
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('median house price')
plt.title("House price VS Year sold")
```

```python
#here we will compare the differance between all years features with saleprice.
data = dataset.copy()
for feature in year_feature:
    if feature!='YrSold':
        data[feature] = data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title('house price according to year')
        plt.show()
        
#below one is power law distribution 
```

# Observations
according to the above plot we can say that new house is more costlier compare to old. one.
it looks like parito or powerlaw distribution.

```python
#numerical variables- 
#2 types..
#continous & discrete variable.
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<=25]
```

```python
print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
#relation ship between discreate and sales price.
data = dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("SalPrice")
    plt.title(feature)
    plt.show()
```

there is relation ship between discrete features and sales price.

```python
#continous Variable/ features..

continous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature]
print(continous_feature)
```

```python

data = dataset.copy()
for feature in continous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
### Lets analyze the continous values by creating histograms to understand the distribution
data=dataset.copy()
for feature in continous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```
